```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
X=range(0,50,10)
Y=range(0,50,10)
X1=range(10,110,20)
Y1=range(100,0,-20)
```

```python
style.use("ggplot")
plt.plot(X,Y,label="first property")
plt.plot(X1,Y1,label="second property",color='g')
plt.title('Simple XY scatter')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
plt.grid(True,color='k')

plt.show()

```

```python
style.use("ggplot")
plt.bar(X,Y,label="first property")
#plt.plot(X1,Y1,label="second property",color='g')
plt.title('Simple XY scatter')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
plt.grid(True,color='k')

plt.show()
```

```python
style.use("ggplot")
plt.scatter(X,Y,label="first property")
plt.scatter(X1,Y1,label="second property",color='g')
plt.title('Simple XY scatter')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
#plt.grid(True,color='k')

plt.show()
```
