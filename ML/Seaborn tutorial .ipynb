```python
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
%matplotlib inline
import pandas as pd
#https://elitedatascience.com/python-seaborn-tutorial
```

```python
s1=pd.Series(range(0,100,10))
s2=pd.Series(range(5,105,10))
```

```python
pd1=pd.DataFrame({'a':s1,'b':s2})
```

```python
pd1.head()
```

```python
sns.lmplot(x='a', y='b', data=pd1)
```

```python
sns.set_style('whitegrid')
sns.violinplot(x='a', y='a', data=pd1)
```

```python
sns.swarmplot(x='a', y='b', data=pd1)
```

```python
s3=pd.Series(range(0,100,20))
```

```python
pd2=pd.DataFrame({'a':s1,'b':s2,'c':s3})
```

```python
sns.lmplot(x='a', y='b', data=pd2)
```

```python
sns.lmplot(x='a', y='b', data=pd2,fit_reg=False) 
```

```python
sns.boxplot(data=pd2)
```

```python
s4=pd.Series(range(0,100,5))
```

```python
pd3=pd.DataFrame({'a':s1,'b':s2,'c':s3,'d':s4})
sns.lmplot(x='a', y='b', data=pd3,fit_reg=False,hue='d') 
```

```python
corr = pd3.corr()
```

```python
sns.heatmap(corr)
```

```python
s5=pd.Series(range(100,0,-5))
pd4=pd.DataFrame({'a':s1,'b':s2,'c':s3,'d':s4,'e':s5})
corr1 = pd4.corr()
sns.heatmap(corr1)
```

```python
sns.factorplot(x='a', y='b',data=pd4, 
                   hue='d',  # Color by stage
                   col='c',  # Separate by stage
                   kind='swarm')
```

```python
sns.countplot(x='a', data=pd4)
```
